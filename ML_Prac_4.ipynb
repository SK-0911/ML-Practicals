```python
import pandas as pd
df=pd.read_csv("data.csv")
df.head()
```

```python
#drop columns which are not required
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns', inplace=True)
df.head()
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs=pd.concat([inputs, dummies],axis='columns')
inputs.head()
```

```python
inputs.drop('Sex', axis='columns', inplace=True)
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
#Fill all the N/A values with the mean of Age column
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
#Check again if there are any N/A values
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split

#Splitting the training and testing data in 80%/20% ratio
X_train, X_test, y_train, y_test=train_test_split(inputs, target, test_size=0.2)
```

```python
print("Total rows in the dataset: ", str(len(inputs)))
print("Total rows in the training dataset: ", str(len(X_train)))
print("Total rows in the testing dataset: ", str(len(X_test)))
```

```python
#Importing the naive bayes
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
#Accuracy score of the model
model.score(X_test, y_test)
```
